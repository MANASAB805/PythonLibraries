```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.title("1st graph")
plt.ylabel("This is Y axis")
plt.xlabel("This is X axis")
plt.show()
```

```python
plt.plot(x, y)
plt.plot([1,2,3],[2,5,7])
plt.title("1st graph")
plt.ylabel("This is Y axis")
plt.xlabel("This is X axis")
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot(x, y,label="first")
plt.plot([1,2,3],[2,5,7],label="Second")
plt.title("1st graph")
plt.ylabel("This is Y axis")
plt.xlabel("This is X axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
  
year = [1972, 1982, 1992, 2002, 2012]
india = [100.6, 158.61, 305.54, 394.96, 724.79]
bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]
plt.plot(year, india, color ='orange', marker ='o', markersize = 12, label ='India')
plt.plot(year, bangladesh, color ='g',linestyle ='dashed', linewidth = 2,label ='Bangladesh')
plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')
plt.title('Electricity consumption per \ capita of India and Bangladesh')
plt.legend()
plt.show()
```
